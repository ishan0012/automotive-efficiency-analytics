```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# 1. Load the professional dataset directly from Seaborn
df = sns.load_dataset('mpg')

# 2. Quick Clean: Remove rows with missing values (horsepower)
df = df.dropna()

print("Dataset Loaded Successfully! Shape:", df.shape)
df.head()
```

```python
# 1. Interactive Horsepower vs. Mileage (The "Wow" Factor)
fig = px.scatter(df, x="horsepower", y="mpg", size="weight", color="origin",
                 hover_name="name", log_x=True, size_max=60,
                 title="Horsepower vs. Fuel Efficiency (Sized by Weight)")
fig.show()

# 2. Origin Analysis (Comparison)
plt.figure(figsize=(10, 6))
sns.violinplot(x='origin', y='mpg', data=df, palette='Set2', hue='origin', legend=False)
plt.title('Fuel Efficiency Distribution by Region (USA vs Japan vs Europe)')
plt.show()

# 3. Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.select_dtypes(include=['float64', 'int64']).corr(), 
            annot=True, cmap='RdYlGn', fmt=".2f")
plt.title('Engineering Feature Correlations')
plt.show()
```
